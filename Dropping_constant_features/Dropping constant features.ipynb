# Feature Selection - Dropping Constant Features

```python
import pandas as pd

data = pd.DataFrame({"A":[1,2,4,1,2,4],
                    "B":[4,5,6,7,8,9],
                    "C":[0,0,0,0,0,0],
                    "D":[1,1,1,1,1,1]})

data.head()
```

## Variance Threshold:
**This feature selection algorithm looks at the features(X, not the desired outputs(y), and can thus be used for unsupervised learning**  
  
**In easier way it will remove all the features which have the variance 0**

```python
from sklearn.feature_selection import VarianceThreshold

var_thres = VarianceThreshold(threshold=0)
var_thres.fit(data)
```

```python
constant_columns = [column for column in data.columns
                   if column not  in data.columns[var_thres.get_support()]]

print((constant_columns))
```

```python
for features in constant_columns:
    print(features)
```

```python
data.drop(constant_columns,axis=1,inplace=True)
data
```
